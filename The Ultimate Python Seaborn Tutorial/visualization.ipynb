```python
# Import the packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('pokemon.csv', index_col=0)
```

```python
df.head()
```

```python
# Filter out the dataset and include few rows of each Generation
instances = 20
final_df = pd.DataFrame()
for i in range(1, 7):
    final_df = final_df.append(df[df['Generation'] == i][:instances])
```

```python
# Scatter Plot with Regression Line
sns.lmplot(x='Attack', y='Defense', data=final_df)

# Scatter Plot with no Regression Line
# sns.lmplot(x='Attack', y='Defense', data=final_df, fit_reg=False)
```

```python
# Scatter Plot with no Regression line and different hues
sns.lmplot(x='Attack', y='Defense', data=final_df, fit_reg=False, hue='Generation')
```

```python
# Box Plot
# Box Plot Description
# Shows the skewness of the values
# Center line is Median for the feature
# Each quartile is 25% of the values
sns.boxplot(data=final_df.drop(['Generation', 'Legendary'], axis=1))
```

```python
# Strip Plot
sns.stripplot(x="Generation", y="HP", data=final_df, jitter=True)
```

```python
# Swarm Plot
sns.swarmplot(x='Generation', y='HP', data=final_df)
```

```python
# Heatmap
corr = final_df.drop(['Legendary', 'Generation'], axis=1).corr()
sns.heatmap(corr, annot=True)
```

```python
# Distribution Histogram
sns.distplot(final_df.Attack)
```

```python
# Count Bar Chart
sns.countplot(y='Type 1', data=final_df, color='#00ffff')
```

```python
# Factor Plot
g = sns.factorplot(x='Type 1', y='Attack', data=final_df[:instances*3], 
                   hue='Generation', col='Generation', kind='swarm')
g.set_xticklabels(rotation=-45)
```

```python
# Density Plot
sns.kdeplot(final_df['Attack'], final_df['Defense'])
```

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=final_df)
```
